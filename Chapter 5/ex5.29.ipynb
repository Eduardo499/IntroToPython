```python
import seaborn as sns

numbers = [174, 335, 278, 214, 422, 513, 737, 672, 489, 412, 1301, 1105, 1123, 1376, 1502, 894, 665, 1704, 1656, 1342]

title = 'Infected patients for each day'
axes = sns.barplot(x=list(range(1, len(numbers) + 1)), y=numbers)

axes.set_title(title)
axes.set(xlabel='Days', ylabel='Number of patients')
```
