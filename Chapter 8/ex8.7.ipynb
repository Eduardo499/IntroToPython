```python
def display_character_table():
    print(f"{'Code':<10}{'Character':<10}")
    print("-" * 20)

    for code in range(256):
        try:
            char = chr(code)
            if char.isprintable():
                print(f"{code:<10}{char:<10}")
            else:
                print(f"{code:<10}{'N/A':<10}")
        except ValueError:
            print(f"{code:<10}{'N/A':<10}")

if __name__ == '__main__':
    display_character_table()
```
