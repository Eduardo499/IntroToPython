```python
def maximum(value1, value2, value3):
    """
    Return the maximum of three values.

    >>> maximum(1, 2, 3)
    3
    >>> maximum(2, 1, 3)
    3
    >>> maximum(2, 3, 1)
    3

    >>> maximum(1.1, 2.2, 3.3)
    3.3
    >>> maximum(2.2, 1.1, 3.3)
    3.3
    >>> maximum(2.2, 3.3, 1.1)
    3.3

    >>> maximum("apple", "banana", "cherry")
    'cherry'
    >>> maximum("banana", "apple", "cherry")
    'cherry'
    >>> maximum("banana", "cherry", "apple")
    'cherry'
    """
    max_value = value1
    if value2 > max_value:
        max_value = value2
    if value3 > max_value:
        max_value = value3
    return max_value

import doctest
doctest.testmod()
```
