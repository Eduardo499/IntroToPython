```python
import random

def roll_dice():
    """Roll two dice and return their face values as a tuple."""
    die1 = random.randrange(1, 7)
    die2 = random.randrange(1, 7)
    return (die1, die2) # pack die face values into a tuple
    
def play_game():
    """Play the game"""
    rolls = 0
    die_values = roll_dice() # first roll
    rolls += 1
    
    # determine game status and point, based on first roll
    sum_of_dice = sum(die_values)
    
    if sum_of_dice in (7, 11): # win
        return 'WON', rolls
    elif sum_of_dice in (2, 3, 12): # lose
        return 'LOST', rolls
    else: # remember point
        my_point = sum_of_dice
        
    # continue rolling until player wins or loses
    while True:
        die_values = roll_dice()
        rolls += 1
        sum_of_dice = sum(die_values)
        
        if sum_of_dice == my_point: # win by making point
            return 'WON', rolls
        elif sum_of_dice == 7: # lose by rolling 7
            return 'LOST', rolls

# Main program loop to play multiple times
num_games = 1000000
wins = 0
losses = 0
roll_stats = {}

for _ in range(num_games):
    result, rolls = play_game()
    
    if result == 'WON':
        wins += 1
    else:
        losses += 1

    if rolls not in roll_stats:
        roll_stats[rolls] = {'WON': 0, 'LOST': 0}
    roll_stats[rolls][result] += 1

print(f'Porcentage of wins: {(wins / (wins + losses) * 100):.2f}%')
print(f'Porcentage of losses: {(losses / (wins + losses) * 100):.2f}%')
print('Percentage of wins/losses based on total number of rolls')

print()

cumulative_games = 0

for rolls in sorted(roll_stats):
    total_games = roll_stats[rolls]['WON'] + roll_stats[rolls]['LOST']
    cumulative_games += total_games
    roll_win_percentage = (roll_stats[rolls]['WON'] / num_games) * 100
    roll_loss_percentage = (roll_stats[rolls]['LOST'] / num_games) * 100
    cumulative_percentage = (cumulative_games / num_games) * 100

    print(f"Rolls: {rolls}, Won: {roll_win_percentage:.2f}%, Lost: {roll_loss_percentage:.2f}%, Cumulative: {cumulative_percentage:.2f}%")
    
```
