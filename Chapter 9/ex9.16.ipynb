```python
import pandas as pd
import matplotlib.pyplot as plt

def main():
    # Load the dataset
    file_path = 'diamonds.csv'  # Replace with the correct file path
    df = pd.read_csv(file_path, index_col=0)
    
    # Check and print column names to find the cause of KeyError
    print("Columns in the dataset:", df.columns)

    # Optionally, strip any extra spaces from column names
    df.columns = df.columns.str.strip()

    # Display first rows to inspect data
    print("\nFirst few rows of the DataFrame:")
    print(df.head())

    # Proceed with the rest of the steps...
    print("\nDescriptive statistics for numerical columns:")
    print(df.describe())

    # Handling categorical columns
    print("\nDescriptive statistics for categorical columns:")
    if 'cut' in df.columns:
        print("Column 'cut':")
        print(df['cut'].describe())
    else:
        print("Column 'cut' not found.")
    
    if 'color' in df.columns:
        print("\nColumn 'color':")
        print(df['color'].describe())
    else:
        print("Column 'color' not found.")
    
    if 'clarity' in df.columns:
        print("\nColumn 'clarity':")
        print(df['clarity'].describe())
    else:
        print("Column 'clarity' not found.")
    
    # Generate histograms
    print("\nGenerating histograms for the numerical columns...")
    df.hist(figsize=(10, 8), bins=20, edgecolor='black', grid=False)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```
