```python
import pandas as pd
import matplotlib.pyplot as plt

def main():
    # Load the Iris dataset
    df = pd.read_csv('iris.csv', index_col=0)

    # Display the first 5 rows of the DataFrame
    print("First 5 rows of the DataFrame:")
    print(df.head())

    # Display the last 5 rows of the DataFrame
    print("\nLast 5 rows of the DataFrame:")
    print(df.tail())

    # Descriptive statistics for numerical columns
    print("\nDescriptive statistics for numerical columns:")
    print(df.describe())

    # Enable matplotlib support in IPython (only needed in Jupyter/IPython environments)
    %matplotlib inline

    # Generate histograms for each numerical column
    print("\nGenerating histograms for numerical columns...")
    df.hist(figsize=(10, 8), bins=20, edgecolor='black', grid=False)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```
