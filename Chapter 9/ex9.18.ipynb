```python
import pandas as pd

# Load the Anscombe's Quartet dataset
url = 'https://gist.githubusercontent.com/ericbusboom/b2ac1d366c005cd2ed8c/raw/c92c66e43d144fa9c29dbd602d5af6988e8db533/anscombes.csv'
df = pd.read_csv(url)

# Display the first 5 rows of the dataset
print(df.head())
```

```python
import matplotlib.pyplot as plt

# Scatter plot for Anscombe's Quartet
# We will plot each of the four datasets (dataset 1, 2, 3, and 4) in different subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Loop through each dataset in the quartet and plot
for i, dataset in enumerate(df['dataset'].unique()):
    ax = axes[i//2, i%2]  # Select subplot position
    subset = df[df['dataset'] == dataset]
    ax.scatter(subset['x'], subset['y'])
    ax.set_title(f"Dataset {dataset}")
    ax.set_xlabel('x')
    ax.set_ylabel('y')

# Show the plot
plt.tight_layout()
plt.show()
```
